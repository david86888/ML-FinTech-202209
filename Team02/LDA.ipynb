## Latent Dirichlet Allocation(LDA)
1.建立Corpus，每個token對應一個token’s id。

```python
dictionary_LDA = corpora.Dictionary(sentence_token) 
print(dictionary_LDA.token2id)
```

．大量文檔可以利用這個方式縮減數據，因為dictionary_LDA只有300多個詞，不應用。

```python
#dictionary_LDA.filter_extremes(no_below=3)
#print(dictionary_LDA.token2id)
```

2.建立以一則訊息為單位的Corpus，形式為(token_id, token_count)。

```python
corpus = [dictionary_LDA.doc2bow(list_of_tokens) for list_of_tokens in sentence_token]
corpus
```

```python
num_topics = 3
lda_model = models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary_LDA, passes=4, \
                alpha=[0.01]*num_topics, eta=[0.01]*len(dictionary_LDA.keys()))
for i,topic in lda_model.show_topics(formatted=True, num_topics=num_topics, num_words=7):
    print(str(i)+": "+ topic)
    print()
```

```python
#dictionary_LDA.filter_extremes(no_below=3)
passes = [ 20, 25, 30, 35 ]
for num in passes:
    print( 'passes=', num, sep='')
    num_topics = 3
    lda_model = models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary_LDA, passes=num, \
                                alpha=[0.01]*num_topics, eta=[0.01]*len(dictionary_LDA.keys()))
    for i,topic in lda_model.show_topics(formatted=True, num_topics=num_topics, num_words=10):
        print(str(i)+": "+ topic)
        print()
    print()
```

```python
alpha = [ 0.01, 0.05, 0.1, 0.5, 1 ]
for num in alpha:
    print( 'alpha=', num, sep='')
    num_topics = 3
    lda_model = models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary_LDA, passes=25, \
                                alpha=[num]*num_topics, eta=[0.01]*len(dictionary_LDA.keys()))
    for j,topic in lda_model.show_topics(formatted=True, num_topics=num_topics, num_words=10):
        print(str(i)+": "+ topic)
        print()
    print()
```
