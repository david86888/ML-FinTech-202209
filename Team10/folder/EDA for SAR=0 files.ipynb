```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Dataset : ccbaNOSAR

- cust_id : 顧客編號 
- lupay : 上月繳款額度 (numerical)
- byymm : 帳務年月 (categorical)
- cycam : 信用額度 (numerical)
- usgam : 已使用額度 (numerical)
- clamt : 本月分期預借現金金額 (numerical)
- csamt : 本月預借現金金額 (numerical)
- inamt : 本月分期消費金額 (numerical)
- cucsm : 本月消費金額 (numerical)
- cucah : 本月借現金額 (numerical)

```python
ccbaNOSAR = pd.read_excel('ccbaNOSAR.xlsx')
ccbaNOSAR.head()
```

```python
# Check the numbers of rows and columns
ccbaNOSAR.shape
```

```python
# Check if the NA values exist
ccbaNOSAR.isnull().sum()
```

```python
ccbaNOSAR.describe()
```

```python
# Display all columns
pd.set_option("display.max_rows", None)
ccbaNOSAR['byymm'].value_counts()
```

```python
# Draw the barplots for categorical variables
ccbaNOCategorical = ['byymm']
for col in ccbaNOCategorical:
    ccbaNOSAR[col].value_counts().plot.bar(title='Freq dist of '+ col)
    plt.show()
```

```python
# Draw the histograms for numerical variables
ccbaNONumerical = ['lupay','cycam','usgam','clamt','csamt','inamt','cucsm','cucah']
for col in ccbaNONumerical:
    ccbaNOSAR[col].hist(figsize = (6,6), bins = 20)
    plt.title('Histogram for ' + col)
    plt.show()
    
# 可以看得出來變數都是高峰在於0附近
```

```python
# Show the scatter plots of variables
sns.pairplot(ccbaNOSAR[['lupay','cycam','usgam','clamt','csamt','inamt','cucsm','cucah']])
plt.show()
```

```python
# Show the heatmap and the relationship of correlation
plt.figure(figsize=(8,8))
sns.heatmap(ccbaNOSAR.corr(), annot = True, center=0.0, cmap='coolwarm');
#sns.heatmap(dataset.corr(), annot = False, center=0.0, cmap='YlGnBu'); # cmap="YlGnBu",
plt.show()
```

### Dataset : cdtxSAR

- cust_id : 顧客編號
- date : 消費日期 (Categorical)
- country : 消費地國別，130＝台灣 (Categorical)
- cur_type : 消費地幣別，47＝台幣 (Categorical)
- amt : 交易金額-台幣 (Numerical)

```python
cdtxNOSAR = pd.read_excel('cdtxNOSAR.xlsx')
cdtxNOSAR.head()
```

```python
cdtxNOSAR.shape
```

```python
cdtxNOSAR.isnull().sum()
```

```python
cdtxNOSAR.describe()
```

```python
# Display all columns
pd.set_option("display.max_rows", None)
cdtxNOSAR['date'].value_counts()
```

```python
cdtxNOSAR['country'].value_counts(normalize = True)
# 原始資料中，國別134大概佔資料10%
# 而分開後的有洗錢名單資料，國別134佔的比例高達54%
# 而分開後的沒洗錢名單資料，國別134佔的比例低至1%
# 因此國別可能是很重要的判斷依據之一
```

```python
cdtxNOSAR['cur_type'].value_counts(normalize = True)
```

```python
# Draw the barplots for categorical variables
cdtxNOCategorical = ['date','country','cur_type']
for col in cdtxNOCategorical:
    cdtxNOSAR[col].value_counts().plot.bar(title='Freq dist of '+ col)
    plt.show()
    
# 在幣別變數的部分，我們可以知道原始資料是台幣47最多
# 但在分開後的洗錢資料中是幣別46變次數最多的
# 沒有洗錢的幣別則是47最多
# 因此cur_type 應該也是一個需要注意的地方
```

```python
# Draw the histograms for numerical variables
cdtxNONumerical = ['amt']
for col in cdtxNONumerical:
    cdtxNOSAR[col].hist(figsize = (6,6), bins = 20)
    plt.title('Histogram for ' + col)
    plt.show()
    
# 幾乎都集中在值很小的周圍，且最大不超過250,000(二十五萬左右)
# 所以其實用消費來洗錢的金額通常不會很大，偏向小額多筆
```

```python
cdtxNOSAR['amt'].value_counts()
```

```python
# Show the scatter plots of numerical variables
sns.pairplot(cdtxNOSAR[['date','country','cur_type','amt']])
plt.show()

# amt(交易金額)和country(消費地國家)、cur_type(幣別)都是低正相關性
```

```python
# Show the heatmap and the relationship of correlation
plt.figure(figsize=(8,8))
sns.heatmap(cdtxNOSAR.corr(), annot = True, center=0.0, cmap='coolwarm');
#sns.heatmap(dataset.corr(), annot = False, center=0.0, cmap='YlGnBu'); # cmap="YlGnBu",
plt.show()
```
