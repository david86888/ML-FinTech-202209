```python
# %load predict_trial.py
#!/usr/bin/env python3
"""
Created on Wed Nov 30 18:09:48 2022

@author: siusiu
"""

#%%
import scipy.optimize
import random as rd
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#%%

df = pd.read_csv("merge_y_alertkey.csv")
df2 = pd.read_csv("encode_data.csv")
#%%

df = df.sort_values(['cust_id','date'],ascending=[True,True])
#%%
df2 = df2.sort_values(['cust_id','date'],ascending=[True,True])

#%%

result = df.merge(df2, how='outer',indicator=True).loc[lambda x : x['_merge'] == 'left_only']
result2 = df.merge(df2, how='outer',indicator=True).loc[lambda x : x['_merge'] == 'both']

#%%

result3 = result.merge(result2,how='outer')
#%%
result2 = result2.drop(columns=['_merge'])

#%%
result2.to_csv('informal_trial.csv')

#%%
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree

#%%
X = result2.drop(['alert_key', 'date', 'sar_flag', 'cust_id'],axis=1)
y = result2['sar_flag']

#%%
df3 = pd.read_csv('public_x_alert_date.csv')
#%%
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)

#%%

Logit = LogisticRegression()
Logit_pred = Logit.fit(X_train, y_train).predict(X_test)
#print(Logit_predict_proba(X))
print('\n')
print(confusion_matrix(y_test, Logit_pred))
print(classification_report(y_test, Logit_pred, digits = 3))


#%%
# clf = LogisticRegression(random_state=0).fit(X, y)
# print(clf.predict_proba(X)
#%%
LDA = LinearDiscriminantAnalysis()
LDA_pred = LDA.fit(X_train, y_train).predict(X_test)
print('\n')
print(confusion_matrix(y_test, LDA_pred))
print(classification_report(y_test, LDA_pred, digits = 3))

#%%
QDA = QuadraticDiscriminantAnalysis()
QDA_pred = QDA.fit(X_train, y_train).predict(X_test)
print('\n')
print(confusion_matrix(y_test, QDA_pred))
print(classification_report(y_test, QDA_pred, digits = 3))
#%%
NB = GaussianNB()
NB_pred = NB.fit(X_train, y_train).predict(X_test)
print('\n')
print(confusion_matrix(y_test, NB_pred))
print(classification_report(y_test, NB_pred, digits = 3))

#%%
clf = tree.DecisionTreeClassifier()
tree_train = clf.fit(X_train, y_train)

# 預測
test_y_predicted = tree_train.predict(X_test)
print('\n')
print(confusion_matrix(y_test, test_y_predicted))
print(classification_report(y_test, test_y_predicted, digits = 3))



```
