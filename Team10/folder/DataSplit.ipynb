```python
# %load DataSplit.py
#!/usr/bin/env python3
"""
Created on Wed Oct 19 19:36:24 2022

@author: siusiu
"""

#%%
import scipy.optimize
import random as rd
import pandas as pd
import numpy as np


#%%
ccba = pd.read_csv('public_train_x_ccba_full_hashed.csv') #每月帳戶資訊
cdtx = pd.read_csv('public_train_x_cdtx0001_full_hashed.csv') #消費資訊
custinfo = pd.read_csv('public_train_x_custinfo_full_hashed.csv') #顧客資訊
dp = pd.read_csv('public_train_x_dp_full_hashed.csv') #轉帳交易資訊
remit = pd.read_csv('public_train_x_remit1_full_hashed.csv') #外匯交易資訊

publicAlertDate = pd.read_csv('public_x_alert_date.csv') #警告發生日期
trainAlertDate = pd.read_csv('train_x_alert_date.csv')
trainAnswer = pd.read_csv('train_y_answer.csv')

#%%

alertkey=[]

for i in range(len(trainAlertDate)):
    if(trainAnswer.iloc[i,1]==1):
        alertkey.append(trainAnswer.iloc[i,0])
    else:
        pass
        
#%%
customerID=[]
for i in range(len(custinfo)):
    if(custinfo.iloc[i,0]==alertkey).any():
        customerID.append(custinfo.iloc[i,1])
    else:
        pass

#%%
custSAR=[]
for i in range(len(custinfo)):
    if(custinfo.iloc[i,0]==alertkey).any():
        custSAR.append(custinfo.iloc[i])
    else:
        pass
#%%
customerID = np.array(customerID)

#%%
ccbaSAR=[]
for i in range(len(ccba)):
    if(ccba.iloc[i,0]==customerID).any():
        ccbaSAR.append(ccba.iloc[i])
    else:
        pass
#%%
cdtxSAR=[]
for i in range(len(cdtx)):
    if(cdtx.iloc[i,0]==customerID).any():
        cdtxSAR.append(cdtx.iloc[i])
    else:
        pass
#%%
dpSAR=[]
for i in range(len(dp)):
    if(dp.iloc[i,0]==customerID).any():
        dpSAR.append(dp.iloc[i])
    else:
        pass

#%%
remitSAR=[]
for i in range(len(remit)):
    if(remit.iloc[i,0]==customerID).any():
        remitSAR.append(remit.iloc[i])
    else:
        pass
#%%
cdtxSARdf = pd.DataFrame(data=cdtxSAR)
ccbaSARdf = pd.DataFrame(data=ccbaSAR)
dpSARdf = pd.DataFrame(data=dpSAR)
remitSARdf = pd.DataFrame(data=remitSAR)
custSARdf = pd.DataFrame(data=custSAR)

#%%

cdtxSARdf.to_excel('cdtxSAR.xlsx')
ccbaSARdf.to_excel('ccbaSAR.xlsx')
dpSARdf.to_excel('dpSAR.xlsx')
remitSARdf.to_excel('remitSAR.xlsx')
custSARdf.to_excel('custinfoSAR.xlsx')


#%%

alertkeynoSAR=[]

for i in range(len(trainAlertDate)):
    if(trainAnswer.iloc[i,1]==0):
        alertkeynoSAR.append(trainAnswer.iloc[i,0])
    else:
        pass
        
#%%
customerIDnoSAR=[]
for i in range(len(custinfo)):
    if(custinfo.iloc[i,0]==alertkeynoSAR).any():
        customerIDnoSAR.append(custinfo.iloc[i,1])
    else:
        pass
#%%
customerIDnoSAR = np.array(customerIDnoSAR)
#%%
for i in range(len(customerIDnoSAR)):
    if(customerIDnoSAR[i]==customerID).any():
        customerIDnoSAR = np.delete(customerIDnoSAR,i,0)


#%%
custnoSAR=[]
for i in range(len(custinfo)):
    if(custinfo.iloc[i,0]==alertkeynoSAR).any():
        custnoSAR.append(custinfo.iloc[i])
    else:
        pass
#%%
custnoSAR = np.array(custnoSAR)
for i in range(1,len(custnoSAR)):
    if(custnoSAR[i,1]==customerID).any():
        custnoSAR = np.delete(custnoSAR,i,0)


#%%
ccbanoSAR=[]
for i in range(len(ccba)):
    if(ccba.iloc[i,0]==customerIDnoSAR).any():
        ccbanoSAR.append(ccba.iloc[i])
    else:
        pass
#%%
cdtxnoSAR=[]
for i in range(len(cdtx)):
    if(cdtx.iloc[i,0]==customerIDnoSAR).any():
        cdtxnoSAR.append(cdtx.iloc[i])
    else:
        pass
#%%
dpnoSAR=[]
for i in range(len(dp)):
    if(dp.iloc[i,0]==customerIDnoSAR).any():
        dpnoSAR.append(dp.iloc[i])
    else:
        pass

#%%
remitnoSAR=[]
for i in range(len(remit)):
    if(remit.iloc[i,0]==customerIDnoSAR).any():
        remitnoSAR.append(remit.iloc[i])
    else:
        pass
#%%
cdtxnoSARdf = pd.DataFrame(data=cdtxnoSAR)
ccbanoSARdf = pd.DataFrame(data=ccbanoSAR)
dpnoSARdf = pd.DataFrame(data=dpnoSAR)
remitnoSARdf = pd.DataFrame(data=remitnoSAR)
custnoSARdf = pd.DataFrame(data=custnoSAR)
#%%

cdtxnoSARdf.to_excel('cdtxNOSAR.xlsx')
ccbanoSARdf.to_excel('ccbaNOSAR.xlsx')
dpnoSARdf.to_csv('dpNOSAR.csv')
remitnoSARdf.to_excel('remitNOSAR.xlsx')
custnoSARdf.to_excel('custinfoNOSAR.xlsx')
```
